# Identify missing values

```python
# Check for missing values
df.isnull().sum(),df.dtypes
```

# Fill missing values

```python
# Calculate median values
median_new_cases = df['New_cases'].median()
median_new_deaths = df['New_deaths'].median()

# Display the median values (corrected variable names)
median_new_cases, median_new_deaths  

# Fill missing values with median values
df['New_cases'] = df['New_cases'].fillna(median_new_cases)
df['New_deaths'] = df['New_deaths'].fillna(median_new_deaths)

# Calculate mode values
mode_product_cc = df['Country_code'].mode()[0]
mode_product_region = df['WHO_region'].mode()[0]

# Display the mode values
mode_product_cc, mode_product_region  

# Fill missing values with mode values
df['Country_code'] = df['Country_code'].fillna(mode_product_cc)
df['WHO_region'] = df['WHO_region'].fillna(mode_product_region)
```

# Handle Incorrect Data Types:

```python
# Convert Date column to datetime
df['Date_reported'] = pd.to_datetime(df['Date_reported'])

# Ensure numerical columns are in correct type
df['New_cases'] = pd.to_numeric(df['New_cases'], errors='coerce')
df['New_deaths'] = pd.to_numeric(df['New_deaths'], errors='coerce')
```

# Handle Outliers:

```python
# Boxplot to detect outliers
plt.figure(figsize=(12,5))
sns.boxplot(x=df['Cumulative_cases'])
plt.show()
```

# Remove extreme outliers (using IQR):

```python
# Remove outliers (using IQR)
Q1 = df['New_cases'].quantile(0.25)
Q3 = df['New_cases'].quantile(0.75)
IQR = Q3 - Q1
# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out extreme values
df = df[(df['New_cases'] >= lower_bound) & (df['New_cases'] <= upper_bound)]
```

# Save the Cleaned Dataset:

```python
# Save cleaned data
df.to_csv("WHO-COVID-19-global-data-cleaned.csv", index=False)
```
